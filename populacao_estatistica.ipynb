ESTATISTICA POPULACIONAL (POPULACAO)

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
x = iris.data[:, 0:5]
feature_names = ['sepal length (cm)' , 'sepal width (cm)', 'petal length (cm)','petal width (cm)']
x = pd.DataFrame(x,columns=feature_names)
x
```

```python
x.hist()
plt.show()
```

```python
dataset=pd.read_csv('Iris.csv',sep=',')
dataset =pd.DataFrame(dataset)
dataset
```

```python
media = x['sepal length (cm)'].mean()
print('Media: ', media)
```

```python
mediana = x['sepal length (cm)'].median()
print('mediana: ', mediana)
```

```python
q1 = x['sepal length (cm)'].quantile(0.25)
q2 = x['sepal length (cm)'].quantile(0.5)
q3 = x['sepal length (cm)'].quantile(0.75)
print(q1)
print(q2)
print(q3)
```

```python
sns.boxplot(y='petal width (cm)',data=x)
plt.show()
```

```python
#fazendo tabela
print('Frequência ds_categoria:')
a=pd.value_counts(dataset['Species'])
a = pd.DataFrame(data=a)
a.rename(columns={'Species' : 'fi'}, inplace=True)
soma=a['fi'].sum()
a['fri %']=a['fi']/soma*100
a['fri %']=a['fri %'].round(2)
a
```
